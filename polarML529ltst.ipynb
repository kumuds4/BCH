```
# AWGN Channel Simulation and Decoding

# %% [markdown]
# # Polar Code Simulation - AWGN Channel

# %%
# Import necessary libraries
import logging
import numpy as np
import scipy
import math
import pandas as pd
import sklearn
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, TensorDataset
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import scipy.special as sps
import logging, traceback, sys

# Device Configuration
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"ðŸš€ Using Device: {DEVICE}")

# Configuration
BLOCK_LENGTH = 128
INFO_BITS = 64
NUM_SAMPLES_TRAIN = 140000
NUM_TRIALS_PERF = 2000
EPOCHS = 100
BATCH_SIZE = 32
LEARNING_RATE = 1e-3
SNR_AWGN = np.linspace(0, 5, 11)
# Define channel type specifically for this script
CHANNEL_TYPE = 'AWGN'
# Define SNR for training data generation (can be a single value or mean of the range)
TRAINING_SNR_DB = SNR_AWGN.mean()
# Define the list sizes for SCL comparison (used in plotting labels and SCL execution)
LIST_SIZES = [1, 8, 16]


# %%
# --- Polar Code Generator Class ---
class PolarCodeGenerator:
    def __init__(self, N, K, crc_type='CRC-7'):
        self.N = N
        self.K = K
        self.crc_type = crc_type
        self.crc_polynomials = {
            'CRC-7': {'polynomial': [1, 1, 1, 0, 0, 1, 1], 'length': 7}
        }

    def generate_info_bits(self):
        return np.random.randint(2, size=self.K)

    def compute_crc(self, bits):
        poly_info = self.crc_polynomials.get(self.crc_type)
        if not poly_info:
            raise ValueError(f"Unsupported CRC type: {self.crc_type}")

        polynomial = poly_info['polynomial']
        crc_length = poly_info['length']
        message = bits.tolist() + [0] * crc_length
        for i in range(len(message) - crc_length):
            if message[i] == 1:
                for j in range(crc_length + 1):
                    message[i + j] ^= polynomial[j] if j < len(polynomial) else 0

        return np.array(message[-crc_length:], dtype=int)

    def polar_encode(self, info_bits):
        crc_bits = self.compute_crc(info_bits)
        extended_info_bits = np.concatenate([info_bits, crc_bits])
        codeword = np.zeros(self.N, dtype=int)
        codeword[:len(extended_info_bits)] = extended_info_bits
        return codeword

    def verify_codeword(self, codeword):
        poly_info = self.crc_polynomials[self.crc_type]
        crc_length = poly_info['length']
        info_bits = codeword[:-crc_length]
        received_crc = codeword[-crc_length:]
        computed_crc = self.compute_crc(info_bits)
        return np.array_equal(received_crc, computed_crc)

# %%
# --- Enhanced Channel Simulator Class (for AWGN) ---
class EnhancedChannelSimulator:
    def __init__(self, channel_type='AWGN'):
        self.channel_type = channel_type
        logging.basicConfig(level=logging.INFO, format='%(asctime)s [%(levelname)s]: %(message)s', datefmt='%Y-%m-%d %H:%M:%S')
        logging.info(f"Initializing {channel_type} Channel Simulator")


    def simulate(self, encoded_signal, snr_db):
        try:
            encoded_signal = np.array(encoded_signal, dtype=float)
            # Convert 0/1 to +1/-1 BPSK symbols (or -1/+1 depending on convention)
            # Let's use 0 -> +1, 1 -> -1 (common in communications)
            bpsk_signal = 1 - 2 * encoded_signal
            # Assumed signal power is 1 (since symbol is +1 or -1)
            signal_power = 1.0

            snr_linear = 10 ** (snr_db / 10)
            # SNR_linear = signal_power / noise_power
            # noise_power = signal_power / snr_linear = 1 / snr_linear
            # For real noise (AWGN), noise_power = N0/2 * 2 = N0
            # In the context of the LLR formula 2*y/sigma^2 where sigma^2 is variance per dimension
            # and total noise power is N0 = 2*sigma^2, we have sigma^2 = noise_power / 2
            noise_power = signal_power / snr_linear
            noise_variance = noise_power / 2.0 # Variance in real and imaginary parts (for complex) or just real (for real)
            noise_std = np.sqrt(noise_variance)


            if self.channel_type == 'AWGN':
                # Add real Gaussian noise with mean 0 and standard deviation noise_std
                noise = np.random.normal(0, noise_std, bpsk_signal.shape)
                received_signal = bpsk_signal + noise
            elif self.channel_type == 'Rayleigh':
                 # This part will not be used in the AWGN script but is kept for completeness
                 # A simple Rayleigh fading model (amplitude only)
                 fading = np.random.rayleigh(scale=np.sqrt(signal_power / 2.0), size=bpsk_signal.shape) # Scale so mean squared fading is 1 (E[h^2]=Omega)
                 noise = np.random.normal(0, noise_std, bpsk_signal.shape)
                 # received_signal = fading * bpsk_signal + noise # Simple amplitude fading without phase
                 # For real BPSK, Rayleigh fading is usually complex. Let's keep it real for simplicity here as per the original
                 received_signal = fading * bpsk_signal + noise

            else:
                raise ValueError(f"Unsupported channel type: {self.channel_type}")

            # Return the raw received signal (soft values)
            return received_signal
        except Exception as e:
            logging.error(f"Channel simulation error: {e}")
            traceback.print_exc(file=sys.stdout)
            # Return the original signal on error (or handle as appropriate)
            return 1 - 2 * np.array(encoded_signal, dtype=float) # Convert back to BPSK if possible


    def compute_theoretical_performance(self, block_length, snr_linear):
        try:
            if self.channel_type == 'AWGN':
                # Theoretical BER for BPSK in AWGN
                bep = 0.5 * sps.erfc(np.sqrt(snr_linear))
            elif self.channel_type == 'Rayleigh':
                 # Theoretical BER for BPSK in Rayleigh (assuming perfect CSI)
                 # Formula is 0.5 * (1 - sqrt(SNR_avg / (1 + SNR_avg))) where SNR_avg is E[h^2] * Eb/N0
                 # If E[h^2] = 1, then SNR_avg = Eb/N0 = snr_linear
                 bep = 0.5 * (1 - np.sqrt(snr_linear / (1 + snr_linear)))
            else:
                raise ValueError(f"Unsupported channel type: {self.channel_type}")

            # Theoretical BLER is complex for Polar codes; using a simple bound might be misleading.
            # Using a very loose upper bound (Union Bound)
            bler = 1 - (1 - bep) ** block_length # This is a very loose bound and might not be accurate for polar codes
            return bep, bler
        except Exception as e:
            logging.error(f"Theoretical performance computation error: {e}")
            traceback.print_exc(file=sys.stdout)
            return np.zeros_like(snr_linear), np.ones_like(snr_linear)


# %%
# --- Dataset Preparation Function ---
def prepare_polar_dataset(polar_code_gen, num_samples, snr_db, channel_type):
    channel_simulator = EnhancedChannelSimulator(channel_type=channel_type)
    X, y = [], []

    print(f"Generating {num_samples} samples for {channel_type} channel at {snr_db:.1f} dB...")
    for i in range(num_samples):
        if (i + 1) % 10000 == 0:
            print(f"  Generated {i + 1}/{num_samples} samples...")
        info_bits = polar_code_gen.generate_info_bits()
        encoded_signal = polar_code_gen.polar_encode(info_bits)
        received_signal = channel_simulator.simulate(encoded_signal, snr_db)
        X.append(received_signal)
        y.append(info_bits)

    print("Dataset generation complete.")
    return np.array(X), np.array(y)

# %%
# --- Enhanced RNN Decoder Class ---
class EnhancedRNNDecoder(nn.Module):
    def __init__(self, input_size, output_size):
        super(EnhancedRNNDecoder, self).__init__()
        # This is a simple Feed-Forward network, not an RNN.
        # If you intend to use an RNN (like LSTM or GRU), the architecture needs to be changed.
        # For this example, I'll keep the current structure as it seems to be what was used.
        self.model = nn.Sequential(
            nn.Linear(input_size, 128),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(64, output_size),
            nn.Sigmoid() # Sigmoid for multi-label binary classification
        )

    def forward(self, x):
        # Ensure input is flattened if it has more than 2 dimensions (batch_size, features)
        if x.dim() > 2:
            x = x.view(x.size(0), -1)
        return self.model(x)

# %%
# --- Decoder Trainer Class ---
class DecoderTrainer:
    def __init__(self, model, learning_rate=1e-3):
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.model = model.to(self.device)
        self.criterion = nn.BCELoss() # Binary Cross-Entropy Loss for multi-label
        self.optimizer = optim.Adam(self.model.parameters(), lr=learning_rate, weight_decay=1e-5)
        self.scheduler = optim.lr_scheduler.ReduceLROnPlateau(
            self.optimizer,
            mode='min',
            factor=0.5,
            patience=5
        )
        self.train_losses = []
        self.val_losses = []

    def train(self, X, y, epochs=100, batch_size=32, validation_split=0.2):
        # Ensure y is float for BCELoss
        X_tensor = X.to(self.device)
        y_tensor = y.to(self.device).float()

        dataset = TensorDataset(X_tensor, y_tensor)
        train_size = int((1 - validation_split) * len(dataset))
        val_size = len(dataset) - train_size
        train_dataset, val_dataset = torch.utils.data.random_split(dataset, [train_size, val_size])

        train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
        val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

        print(f"Starting training for {epochs} epochs...")
        for epoch in range(epochs):
            self.model.train()
            total_train_loss = 0
            for i, (batch_X, batch_y) in enumerate(train_loader):
                batch_X = batch_X.to(self.device)
                batch_y = batch_y.to(self.device)

                self.optimizer.zero_grad()
                outputs = self.model(batch_X)
                loss = self.criterion(outputs, batch_y)
                loss.backward()
                self.optimizer.step()
                total_train_loss += loss.item()

                if (i + 1) % 100 == 0:
                     print(f'  Epoch [{epoch+1}/{epochs}], Step [{i+1}/{len(train_loader)}], Loss: {loss.item():.4f}')


            train_loss = total_train_loss / len(train_loader)
            self.train_losses.append(train_loss)

            self.model.eval()
            total_val_loss = 0
            with torch.no_grad():
                for batch_X, batch_y in val_loader:
                    batch_X = batch_X.to(self.device)
                    batch_y = batch_y.to(self.device)
                    outputs = self.model(batch_X)
                    loss = self.criterion(outputs, batch_y)
                    total_val_loss += loss.item()

            val_loss = total_val_loss / len(val_loader)
            self.val_losses.append(val_loss)

            self.scheduler.step(val_loss)
            print(f"Epoch [{epoch+1}/{epochs}], Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}")

        print("Training complete.")
        return self.train_losses, self.val_losses

    def predict(self, X):
        if not isinstance(X, torch.Tensor):
            X = torch.FloatTensor(X)
        if X.dim() > 2:
            X = X.view(X.size(0), -1)
        X = X.to(self.device)
        self.model.eval()
        with torch.no_grad():
            outputs = self.model(X)
        # Apply sigmoid and threshold to get binary predictions
        return (outputs > 0.5).cpu().numpy().astype(int)

# %%
# --- SCL Decoder Class ---
# NOTE: This SCL decoder implementation is a simplified conceptual placeholder.
# A full, functional SCL decoder for Polar codes is complex and involves
# specific bit ordering (information and frozen) and recursive decoding steps.
# You will need to replace this with a proper SCL decoder implementation
# if you intend to use it for accurate performance comparison.
class SCLDecoder:
    def __init__(self, N, K, list_size, crc_poly=None):
        self.N = N
        self.K = K
        self.list_size = list_size
        self.crc_poly = crc_poly
        self.paths = None
        self.path_metrics = None
        # Placeholder: Determine actual frozen set based on code construction
        # For this placeholder, we'll assume the first K bits are info bits
        self.frozen_set = set(range(self.K, self.N))
        self.info_set = sorted(list(set(range(N)) - set(self.frozen_set)))


    def _calculate_path_metric(self, llr, decision):
        # Using negative log-likelihood for metric. Lower is better.
        # This is based on LLR = log(P(y|x=0) / P(y|x=1))
        # P(y|x=0) = exp(llr) * P(y|x=1)
        # P(y|x=1) = 1 / (1 + exp(llr)) * P(y) (approx)
        # P(y|x=0) = exp(llr) / (1 + exp(llr)) * P(y) (approx)
        # Metric for decision 0 (x=0) is -log(P(y|x=0)) approx -log(exp(llr)/(1+exp(llr))) = - (llr - log(1+exp(llr))) = log(1+exp(llr)) - llr
        # Metric for decision 1 (x=1) is -log(P(y|x=1)) approx -log(1/(1+exp(llr))) = log(1+exp(llr))
        # A more common metric is based on the log-likelihood ratio
        if decision == 0:
             # Cost for deciding 0 when LLR is llr
             return 0 if llr >= 0 else abs(llr) # Cost is 0 if llr favors 0, otherwise it's the "wrong" magnitude
        else: # decision 1
             # Cost for deciding 1 when LLR is llr
             return abs(llr) if llr >= 0 else 0 # Cost is llr if llr favors 1, otherwise it's the "wrong" magnitude

        # Another common metric: absolute value of LLR if decision is wrong, 0 otherwise.
        # This requires comparing the decision to the sign of the LLR.
        # if (decision == 0 and llr < 0) or (decision == 1 and llr > 0):
        #     return abs(llr)
        # else:
        #     return 0


    def decode(self, received_signal, snr_db):
        # --- Initialization ---
        self.paths = [np.zeros(self.N, dtype=int) for _ in range(self.list_size)]
        self.path_metrics = [0.0 for _ in range(self.list_size)]

        # Convert received signal to LLRs (AWGN specific LLR calculation)
        # Assumes BPSK: 0 -> +1, 1 -> -1
        # Received signal y = (1 - 2*x) + noise, where x is 0 or 1
        # LLR = log( P(y|x=0) / P(y|x=1) )
        # For AWGN, P(y|x) is a Gaussian distribution.
        # LLR = 2 * y / sigma_n^2, where sigma_n^2 is the noise variance
        # Assuming unit signal power (A=1), Eb = 1. SNR = Eb/N0 = 1/N0. N0 = 1/SNR.
        # noise_variance sigma_n^2 = N0 / 2 = 1 / (2 * SNR_linear)
        # LLR = 2 * y / (1 / (2 * SNR_linear)) = 4 * y * SNR_linear. This seems off.
        # Let's re-verify the LLR for BPSK in AWGN with 0->+1, 1->-1 and noise variance sigma^2
        # LLR(x=0 vs x=1 | y) = log(p(y|x=0) / p(y|x=1))
        # p(y|x=0) = (1/sqrt(2*pi*sigma^2)) * exp(-(y - 1)^2 / (2*sigma^2))
        # p(y|x=1) = (1/sqrt(2*pi*sigma^2)) * exp(-(y - (-1))^2 / (2*sigma^2)) = (1/sqrt(2*pi*sigma^2)) * exp(-(y + 1)^2 / (2*sigma^2))
        # LLR = -(y-1)^2/(2*sigma^2) + (y+1)^2/(2*sigma^2)
        # LLR = ( -(y^2 - 2y + 1) + (y^2 + 2y + 1) ) / (2*sigma^2)
        # LLR = (4y) / (2*sigma^2) = 2y / sigma^2
        # Where sigma^2 is the variance of the real-valued noise samples.
        # From EnhancedChannelSimulator: noise_variance = noise_power / 2.0 = (signal_power / snr_linear) / 2.0
        # Assuming signal_power = 1, sigma^2 = 1 / (2 * snr_linear).
        # So, LLR = 2 * y / (1 / (2 * snr_linear)) = 4 * y * snr_linear. This seems correct now based on this derivation.

        snr_linear = 10**(snr_db / 10)
        # Ensure the simulator calculates noise_variance correctly and use that here.
        # Based on the simulator's noise_std = sqrt(noise_power / 2.0), noise_variance = noise_power / 2.0
        # And noise_power = signal_power / snr_linear = 1.0 / snr_linear
        # So noise_variance = 1.0 / (2.0 * snr_linear).
        # LLR = 2 * received_signal / noise_variance = 2 * received_signal / (1.0 / (2.0 * snr_linear)) = 4 * received_signal * snr_linear.

        # However, a more standard form for BPSK in AWGN LLR is 2 * received_signal * Eb/N0
        # If signal amplitude is A, symbol energy Eb = A^2. If A=1, Eb=1.
        # SNR = Eb/N0. N0 = Eb/SNR = 1/SNR.
        # LLR = 2 * received_signal * (Eb / N0) = 2 * received_signal * SNR_linear.
        # Let's stick to the more common LLR = 2 * y * SNR_linear form for now, assuming unit energy.

        llrs = 2 * np.array(received_signal) * snr_linear
        current_llrs = [llrs.copy() for _ in range(self.list_size)]

        # --- Simplified Iterative SCL Decoding (Conceptual) ---
        # This iterative loop is a placeholder for the actual SCL logic which is recursive
        # and processes bits in a specific (polar) order.
        # A proper SCL implementation involves functions like G (bit calculation) and F (LLR update).

        # Simplified loop for bit processing (DOES NOT represent actual SCL bit order)
        for bit_index in range(self.N):
            new_paths = []
            new_path_metrics = []
            new_llrs = [] # Placeholder for updated LLRs for the next stage

            # Expand each current path
            for path_idx in range(len(self.paths)):
                current_path = self.paths[path_idx]
                current_metric = self.path_metrics[path_idx]
                path_llrs = current_llrs[path_idx]

                if bit_index in self.frozen_set:
                    # Frozen bit - decision is typically 0
                    decision = 0
                    new_path = current_path.copy()
                    new_path[bit_index] = decision
                    # Add metric contribution for this decision
                    new_metric = current_metric + self._calculate_path_metric(path_llrs[bit_index], decision)
                    # In a real SCL, LLRs would be updated here based on the decision and previous LLRs
                    new_llrs_for_path = path_llrs.copy() # Placeholder

                    new_paths.append(new_path)
                    new_path_metrics.append(new_metric)
                    new_llrs.append(new_llrs_for_path)

                else:
                    # Information bit - try both 0 and 1
                    for decision in [0, 1]:
                        new_path = current_path.copy()
                        new_path[bit_index] = decision
                        new_metric = current_metric + self._calculate_path_metric(path_llrs[bit_index], decision)
                         # In a real SCL, LLRs would be updated here
                        new_llrs_for_path = path_llrs.copy() # Placeholder

                        new_paths.append(new_path)
                        new_path_metrics.append(new_metric)
                        new_llrs.append(new_llrs_for_path)

            # --- List Management (Sort and Prune) ---
            # Sort paths by metric (lower is better) and keep the best `list_size`
            sorted_indices = np.argsort(new_path_metrics)
            self.paths = [new_paths[i] for i in sorted_indices[:self.list_size]]
            self.path_metrics = [new_path_metrics[i] for i in sorted_indices[:self.list_size]]
            current_llrs = [new_llrs[i] for i in sorted_indices[:self.list_size]] # Keep corresponding LLRs

        # --- CRC Verification and Final Selection ---
        best_path_index = -1
        best_metric = float('inf')

        for i in range(self.list_size):
            candidate_codeword = self.paths[i]
            # Extract the K information bits based on the info_set indices
            candidate_info_bits = candidate_codeword[self.info_set]

            if self.crc_poly: # Check if CRC is used
                 # You would need a method to compute CRC for the extracted info bits
                 # and compare it to the CRC bits in the received codeword (at known positions)
                 # computed_crc = self._compute_crc_for_info_bits(candidate_info_bits)
                 # received_crc = candidate_codeword[crc_bit_indices] # Need to know CRC bit indices
                 # if np.array_equal(computed_crc, received_crc):
                 #    if self.path_metrics[i] < best_metric:
                 #        best_metric = self.path_metrics[i]
                 #        best_path_index = i
                 pass # Placeholder for CRC check
            else: # No CRC used, just select the path with the best metric
                if self.path_metrics[i] < best_metric:
                    best_metric = self.path_metrics[i]
                    best_path_index = i

        # If no path passed CRC (and CRC is used), or no paths generated, fallback to best metric path
        if best_path_index == -1:
             best_path_index = np.argmin(self.path_metrics)

        # Extract and return the decoded information bits from the selected path
        decoded_info_bits = self.paths[best_path_index][self.info_set]
        return decoded_info_bits

    def _compute_crc_for_info_bits(self, info_bits):
        # Placeholder: Implement CRC computation based on self.crc_poly
        pass


# %%
# --- Run SCL Decoder Function ---
def run_scl_decoder(polar_code_gen, SNRS, list_size, channel_type, num_trials):
        results = []
        print(f"Running SCL decoder with List Size {list_size} for {channel_type} channel...")
        for snr_db in SNRS:
            # Generate performance data for each SNR point
            X, y = prepare_polar_dataset(polar_code_gen, num_samples=num_trials, snr_db=snr_db, channel_type=channel_type)

            decoder = SCLDecoder(N=polar_code_gen.N, K=polar_code_gen.K, list_size=list_size)

            decoded_bits = np.array([decoder.decode(x, snr_db) for x in X]) # Pass snr_db to decode

            # Calculate BER and BLER
            # Ensure shapes are compatible for comparison
            if decoded_bits.shape != y.shape:
                 print(f"Warning: Decoded bits shape {decoded_bits.shape} does not match actual bits shape {y.shape}. Skipping performance calculation for SNR {snr_db} dB.")
                 ber = np.nan # Assign NaN if shapes don't match
                 bler = np.nan
            else:
                bit_errors = np.sum(np.abs(decoded_bits - y))
                total_bits = num_trials * polar_code_gen.K
                ber = bit_errors / total_bits if total_bits > 0 else np.nan

                block_errors = np.sum(np.any(decoded_bits != y, axis=1))
                bler = block_errors / num_trials if num_trials > 0 else np.nan

            results.append({'SNR': snr_db, 'BER': ber, 'BLER': bler})
            print(f"SCL Decoder - SNR: {snr_db:.1f} dB, List Size: {list_size}, BER: {ber:.4f}, BLER: {bler:.4f}")

        return results

# %%
# --- Performance Comparison Function (for ML/RNN) ---
def performance_comparison(rnn_trainer, polar_code_gen, snr_range, channel_name, num_trials):
    performance_results = {'BER': [], 'BLER': []}
    channel_simulator = EnhancedChannelSimulator(channel_type=channel_name)

    print(f"Evaluating ML (RNN) decoder performance for {channel_name} channel...")
    for snr_db in snr_range:
        # Generate data for performance evaluation at each SNR point
        X, y = prepare_polar_dataset(polar_code_gen, num_samples=num_trials, snr_db=snr_db, channel_type=channel_name)

        predictions = rnn_trainer.predict(X) # predictions shape: [num_trials, K]
        actual_labels = y # actual_labels shape: [num_trials, K]

        # Calculate BER: Total number of bit errors / Total number of bits
        if predictions.shape != actual_labels.shape:
            print(f"Warning: RNN predictions shape {predictions.shape} does not match actual labels shape {actual_labels.shape}. Skipping performance calculation for SNR {snr_db} dB.")
            ber = np.nan
            bler = np.nan
        else:
            ber = np.sum(np.abs(predictions - actual_labels)) / (num_trials * polar_code_gen.K)

            # Calculate BLER: Number of blocks with at least one bit error / Total number of blocks
            block_errors = np.sum(np.any(predictions != actual_labels, axis=1))
            bler = block_errors / num_trials

        performance_results['BER'].append(ber)
        performance_results['BLER'].append(bler)
        print(f"ML (RNN) Performance - SNR: {snr_db:.1f} dB, BER: {ber:.4f}, BLER: {bler:.4f}")

    return performance_results


# %%
# --- Plotting Functions ---

def plot_losses(train_losses, val_losses, channel_name):
    plt.figure(figsize=(10, 5))
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Validation Loss')
    plt.title(f'{channel_name} Channel - Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid()
    plt.show()

def plot_ber_bler_comparison(ml_results, scl_results_dict, channel_type, snr_range):
    plt.figure(figsize=(12, 10))

    # Plot BER
    plt.subplot(2, 1, 1)
    # Plot ML (RNN) results
    plt.plot(snr_range, ml_results['BER'], label='ML (RNN) Decoder', linestyle='--')

    # Plot SCL results for each list size
    for L, results in scl_results_dict.items():
        # Extract BER values from the list of dictionaries
        scl_ber = [res['BER'] for res in results]
        plt.plot(snr_range, scl_ber, label=f'SCL Decoder (List Size {L})')


    plt.title(f'{channel_type} Channel - BER Performance Comparison')
    plt.xlabel('SNR (dB)')
    plt.ylabel('BER')
    plt.yscale('log')
    plt.ylim(1e-5, 1)
    plt.legend()
    plt.grid(True, which="both", ls="--")

    # Plot BLER
    plt.subplot(2, 1, 2)
    # Plot ML (RNN) results
    plt.plot(snr_range, ml_results['BLER'], label='ML (RNN) Decoder', linestyle='--')

    # Plot SCL results for each list size
    for L, results in scl_results_dict.items():
         # Extract BLER values from the list of dictionaries
        scl_bler = [res['BLER'] for res in results]
        plt.plot(snr_range, scl_bler, label=f'SCL Decoder (List Size {L})')


    plt.title(f'{channel_type} Channel - BLER Performance Comparison')
    plt.xlabel('SNR (dB)')
    plt.ylabel('BLER')
    plt.yscale('log')
    plt.ylim(1e-5, 1)
    plt.legend()
    plt.grid(True, which="both", ls="--")

    plt.tight_layout()
    plt.show()

# %%
# --- Main Execution Block for AWGN ---
iif __name__ == "__main__":
```
