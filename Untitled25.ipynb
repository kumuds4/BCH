```python
import numpy as np
import pandas as pd
from scipy.integrate import solve_ivp

# Physical and environmental parameters
m = 0.95            # Mass (kg)
v0 = 45.0           # Initial launch speed (m/s)
theta = np.radians(35)  # Launch angle converted to radians
A = 0.0043          # Cross-sectional area (m^2)
Cd = 0.47           # Drag coefficient
rho = 1.20          # Air density (kg/m^3)
g = 9.81            # Acceleration due to gravity (m/s^2)
ay = 7.0            # Lateral acceleration from spin (m/s^2)
az = -24.0          # Vertical acceleration from spin (m/s^2)

# Calculate quadratic drag coefficient (gamma)
gamma = 0.5 * Cd * rho * A

# Initial State Vector: [x, y, z, vx, vy, vz]
initial_state = [
    0.0,
    0.0,
    0.0,
    v0 * np.cos(theta),
    0.0,
    v0 * np.sin(theta)
]

def equations_of_motion(t, state):
    """
    Computes derivatives for 3D projectile motion with quadratic drag.
    State vector: [x, y, z, vx, vy, vz]
    """
    x, y, z, vx, vy, vz = state
    speed = np.sqrt(vx**2 + vy**2 + vz**2)

    # Derivatives: [dx/dt, dy/dt, dz/dt, dvx/dt, dvy/dt, dvz/dt]
    return [
        vx,
        vy,
        vz,
        -(gamma / m) * speed * vx,
        ay - (gamma / m) * speed * vy,
        -g + az - (gamma / m) * speed * vz
    ]

def hit_ground_event(t, state):
    """Event function to stop integration when the z-coordinate hits 0."""
    return state[2]

# Configure the event to terminate simulation on down-crossing (z = 0)
hit_ground_event.terminal = True
hit_ground_event.direction = -1

# Solve the system of ODEs using the explicit RK45 Runge-Kutta method
solution = solve_ivp(
    fun=equations_of_motion,
    t_span=[0, 10],
    y0=initial_state,
    method='RK45',
    events=hit_ground_event,
    rtol=1e-8,
    atol=1e-8
)

# Extract landing coordinates and time
t_land = solution.t_events[0][0]
state_land = solution.y_events[0][0]
x_land, y_land, z_land = state_land[0], state_land[1], state_land[2]
vx_land, vy_land, vz_land = state_land[3], state_land[4], state_land[5]

# Format and display the results beautifully in Colab
results_data = {
    "Parameter": [
        "Landing Time",
        "X Position (East)",
        "Y Position (South)",
        "Z Position (Vertical)",
        "Final Speed"
    ],
    "Value": [
        f"{t_land:.4f} s",
        f"{x_land:.4f} m",
        f"{y_land:.4f} m",
        f"{z_land:.4f} m",
        f"{np.sqrt(vx_land**2 + vy_land**2 + vz_land**2):.4f} m/s"
    ]
}

df_results = pd.DataFrame(results_data)
print("=== RUNGE-KUTTA (RK45) TRAJECTORY RESULTS ===")
display(df_results)
```
